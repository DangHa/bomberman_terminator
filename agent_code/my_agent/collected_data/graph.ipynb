```python
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
plt.figure()

data = []

with open('collected_data/weight_coin_001.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')

    for row in plots:
        data.append([float(i) for i in row])

data = np.asarray(data)


plt.plot(data[:,1], color = 'red', label = "runs_into_wall_crate")
plt.plot(data[:,3], color = 'blue',label = "runs_towards_closest_coin_but_not_wall_or_crate")
plt.plot(data[:,9], color = 'orange', label = "get_out_of_bomb_range")
plt.plot(data[:,12], color = 'green', label = "runs_into_explosion")
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')


plt.xlabel('The number of step')
plt.title('The change of weights with epsilon = 0.01')
plt.show()
# 2 changes 2 not changes (get the name also)
```

```python
plt.figure()

win = []
point = []
survive = []

with open('collected_data/win_rate.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')
    
    w = 0
    p = 0
    s = 0
    for idx, row in enumerate(plots):
        if row[0] == "1":
            w += 1 
        if row[1] == "1":
            p += 1 
        if row[2] == "1":
            s += 1 

        if idx == 0:
            win.append(w/1)
            point.append(p/1)
            survive.append(s/1)
        
        if idx == 9:
            win.append(w/10)
            point.append(p/10)
            survive.append(s/10)

        if idx == 99:
            win.append(w/100)
            point.append(p/100)
            survive.append(s/100)

        if idx == 249:
            win.append(w/250)
            point.append(p/250)
            survive.append(s/250)

        if idx == 499:
            win.append(w/500)
            point.append(p/500)
            survive.append(s/500)

x = [1,10,100,250,500]
print(win)
plt.plot(x, win, color = 'red', label = "Winning rate")
plt.plot(x, point, color = 'blue',label = "Leading rate")
# plt.plot(x, survive, color = 'orange', label = "Survived rate")
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')

plt.xlabel('The number of game')
# plt.title('The change of weights with epsilon = 0.95')
plt.show()
```
